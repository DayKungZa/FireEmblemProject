```python
#character pair up with weapon stats
#0 Name
#1 HP
#2 Dmg
#3 Hit
#4 Crit
#5 Avo
#6 Dodge
#7 Spd
#8 Def
#9 Res
#10 HPcost

pairedRef = ["Name","HP","Dmg","Hit","Crit","Avoid","Dodge","Speed","Def","Res","HPcost"];
preBattleRef = ["Name","HP","Dmg","HIT","CRIT","Double?","HPcost"]

def pair(chr, wpn):
    paired = [chr[0] + " + " + wpn[0]]
    paired.append(chr[1])
    paired.append(chr[2] + wpn[1])
    paired.append(chr[3] + wpn[2])
    paired.append((chr[3]+chr[5])//2 + wpn[3])
    paired.append(chr[4] - wpn[4])
    paired.append(chr[5]//2)
    paired.append(chr[4] + wpn[4])
    paired.append(chr[6])
    paired.append(chr[7])
    paired.append(wpn[5])
    return paired

def preBattle(py1, py2):
    p1 = [py1[0]]
    p1.append(py1[1])
    if(py1[10]>0):
        p1.append(py1[2]-py2[9])
    else:
        p1.append(py1[2]-py2[8])
    p1.append(py1[3]-py2[5])
    p1.append(py1[4]-py2[6])
    if(py1[7]-py2[7]>=4):
        p1.append(2)
    else:
        p1.append(1)
    p1.append(py1[10])
    return p1
```
